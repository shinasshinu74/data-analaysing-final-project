```python
#replaced missing values 
df['education'] = df['education'].fillna('Unknown')
df['previous_year_rating'] = df['previous_year_rating'].fillna('Unknown')
```

```python
#removed duplicates
df.drop_duplicates()
```

```python
#confirming that there is no missing valuues
df.isnull().sum()
```

```python
#creating bins and labels for Derived Features
bins = [0, 18, 35, 50, 100]  
labels = ['Teen', 'Young Adult', 'Middle Age', 'Old']
# Create new column AgeGroup
df['AgeGroup'] = pd.cut(df['age'], bins=bins, labels=labels)
```

```python
#transformation
df['gender']=df['gender'].str.upper() #uppercase the letters
```

```python
#outliers
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(5,3))
    sns.boxplot(x=df[col])
    plt.title(f"outlier of {col}")
    plt.show()
```

#EDA & Visualizations

#Univariate Analysis

```python
#histogram
#age distribution
sns.histplot(df['age'], bins=20, kde=True)
plt.title("Distribution of Employee Age")
plt.show()
```

```python
#Age is mostly distributed between 28-34 years
```

```python
# Department count
sns.countplot(x='department', data=df, palette='Set2')
plt.title("Employee Count per Department")
plt.xticks(rotation=45)
plt.show()
```

```python
#sales&marketing departments have higher employee counts.
```

```python
# Boxplot for training score
sns.boxplot(y=df['avg_training_score'], color='orange')
plt.title("Boxplot of Avg Training Score")
plt.show()
```

```python
#Training scores show some outliers
#training_score 40 is on 25%
#training_score 60 is on 50%
#training_score 99 is on 75%
```

#Bivariate Analysis

```python
# Age vs Training Score
sns.scatterplot(x='age', y='avg_training_score', hue='gender', data=df)
plt.title("Age vs Training Score by Gender")
plt.show()
```

```python
#Training score decreases with age for 20-23
#Training score increases with age for 27-53
```

```python
# Department vs Training Score
sns.boxplot(x='department', y='avg_training_score', data=df, palette='Set3')
plt.title("Training Score by Department")
plt.xticks(rotation=45)
plt.show()
```

```python
#HR department and finance department show lower training score
#R&D department and analytics department show higher training scores.
```

```python
# Gender distribution across departments
pd.crosstab(df['department'], df['gender']).plot(kind='bar', figsize=(8,5))
plt.title("Gender Distribution by Department")
plt.ylabel("Count")
plt.show()
```

```python
#male emplyoyees are more than female employees in all department
```

#Multivariate Analysis

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])
# Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm',center=0)
plt.title("Correlation Heatmap (Numeric Columns)")
plt.show()
```

```python
#length of service shows a moderate positive correlation with age
```

#Groupby & Pivot Tables

```python
#groupby
# Average training score by Department
dept_training = df.groupby('department')['avg_training_score'].mean()
print(dept_training)
```

```python
#analytics and r&d departments are in high training score
```

```python
# Pivot: Department vs Gender vs Avg Training Score
pivot_table = df.pivot_table(values='avg_training_score',
                             index='department',
                             columns='gender',
                             aggfunc='mean')
print(pivot_table)
```

```python
#analytics and r&d departments are in high training score and female employees have more traing score than male employees
```

#Combined Visualizations (Subplots)

```python
fig, axes = plt.subplots(1, 2, figsize=(12,5))

# Histogram for Age
sns.histplot(df['age'], bins=20, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title("Age Distribution")

# Boxplot for Training Score
sns.boxplot(y=df['avg_training_score'], ax=axes[1], color='lightgreen')
axes[1].set_title("Training Score Boxplot")

plt.tight_layout()
plt.show()
```

```python
#Subplots allow side-by-side visualization for quick comparisons. Here, we compare age distribution with training score distribution.
```

```python


# Pie chart for the 'department' column
# Count the number of employees in each department
department_counts = df['department'].value_counts()

# Plot pie chart
plt.figure(figsize=(8,8))
plt.pie(department_counts, 
        labels=department_counts.index,   # Names of departments
        autopct='%1.1f%%',               # Show percentages
        startangle=140,)                  # Start angle of first slice
        
plt.title('Employee Distribution by Department')
plt.show()
```

```python
#the highest number of employees are in sales&marketing
#the lowest number of employees are in legal and R&D
```
